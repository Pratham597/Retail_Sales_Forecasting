#Retail Sales Clustering
## Data Cleaning

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# Load dataset
url = ('https://archive.ics.uci.edu/ml/machine-learning-databases'
       '/00352/Online%20Retail.xlsx')
df = pd.read_excel(url, engine='openpyxl')
print('Shape:', df.shape)
df.head()
```

```python
# Basic info
df.info()
print('\nNull counts:\n', df.isnull().sum())
print('\nDuplicates:', df.duplicated().sum())
```

```python
#  Drop nulls & duplicates
df.dropna(subset=['CustomerID'], inplace=True)
df.drop_duplicates(inplace=True)

# Remove cancelled orders (InvoiceNo starts with 'C')
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Remove non-positive Quantity / UnitPrice
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

# Fix dtypes
df['CustomerID'] = df['CustomerID'].astype(int).astype(str)
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])

# Derived column
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

print('Clean shape:', df.shape)
df.describe()
```

```python
#  Save cleaned data
df.to_csv('online_retail_clean.csv', index=False)
print('Saved as online_retail_clean.csv')
```
